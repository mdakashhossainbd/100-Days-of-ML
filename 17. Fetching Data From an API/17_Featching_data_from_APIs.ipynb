```python
#  Shape: 10468*7 for 425 pages
# Need to add by loop. Get and append to the previous dataframe
```

```python
dfs = []

for i in range(1, 425):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=a2244e7e8c17fa0841688cda8404c714&language=en-US&page={i}')
    dfs.append(pd.DataFrame(response.json()['results'])[['id','title','popularity','release_date','vote_average','vote_count','overview']])

df = pd.concat(dfs, ignore_index=True)
```
