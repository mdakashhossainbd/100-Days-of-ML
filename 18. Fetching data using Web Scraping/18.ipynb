# Actual

```
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://poets.org/poems?page=1', headers=headers).text
```

```
soup = BeautifulSoup(webpage, 'lxml')
```

```
print(soup.prettify)
```

```
company=soup.find_all('tr')
```

```
name = []
author = []
year = []

for i in company:
    x = i.find('td', class_='views-field views-field-title')
    x_2 = i.find_all('a', hreflang="und")
    y = i.find('time')

    if x:
        name.append(x.text)
    else:
        name.append("")  # default

    if len(x_2)>1:
        author.append(x_2[1].text)
    else:
        author.append("")  # default

    if y:
        year.append(y.text.strip())
    else:
        year.append("")  # default
```
